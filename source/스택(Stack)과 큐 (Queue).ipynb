## 스택(Stack)의 개념 
#### 스택의 기본 개념을 프로그래밍적인 시각에서 보면 "입력과 출력을 한 방향으로 제한한 알고리즘"이라고 볼 수 있다. 
배열은 배열의 인덱스를 사용하여 배열의 어느 곳이든 액세스가 가능하다. 또한 연결 리스트의 경우에는 링크를 통해서 노드를 검색하고 또 새로운 노드를 연결 리스트의 중간에 삽입하거나 기존의 노드를 삭제하는 것도 가능 했다.
하지만 이러한 방법에는 많은 장점이 있는 반면에 그만큼의 단점도 존재한다.
#### 넣고 뺴는 작업 자체가 복잡하다
연결 리스트를 보면 새로운 노드를 삽입하기 위해서는 기존의 연결 리스트에서 새로운 노드가 삽입될 위치를 검색해야 하고, 또 삽입 과정에도 링크를 연결해야 하므로 간단하지 않다.

#### 이해 비하여 스택의 자료구조는 간단하다
스택이란 바닥부터 데이터를 차곡차곡 쌓는 개념이다.

그릇으로 예시를 들어보자면 첫 번쨰 그릇을 닦아서 두고 두 번쨰 그릇을 닦은 후 첫번쨰 그릇 위에 올려논다 그렇게하다보면 맨마지막의 그릇이 맨위에 올라가게 된다.

1. push : 그릇을 닦아서 천장에 쌓는다
2. pop : 그릇을 사용하기 위해서 가장 위에 있는 그릇을 가지고 온다,
### 이와 같은 방식을 LIFO(Last In FIrst Out)이라고 한다.

    stack.append(item)
    

def pop():
    return stack.pop()

if __name__ == '__main__':
    stack = []
    push(1)
    push(2)
    push(3)
    push(4)
    print("현재 stack의 모습")
    print(stack)
    
    while stack:
        print("POP > {}".format(pop()))
    

## 시간의 효율성
스택 알고리즘은 연결 리스트를 사용하기도 하지만 배열을 사용해서 구혀하기도 한다. 그이유는 스택의 개념 자체가 검색하는 과정이 필요 없어 가장 선두에 있는 데이터를 POP하고 데이터를 삽입할 떄도 가장 선두에 집어 넣기 떄문이다. 따라서 연결 리스트를 사용한다고 하더라도 시간적인 효율성 면에서는 배열보다 나은점이 없다고 본다

## 공간의 효율성
공간의 효율성 측면에서는 배열로 구현한 스택이 배열 전체 크기에 국한된다는 점 떄문에 연결 리스트를 사용한 스택이 더 효율적이라고 말할 수 있다. 그러나 **대부분의 스택을 사용하는 경우에는 스택의 크기를 정해놓고서 사용하는 것이 일반적이기 떄문에 스택을 배열로 구현했다고 해서 공간의 효율성이 저하된다고 보기 어렵다.**

## 코드의 효율성
코드의 효율성은 연결 리스트에 대한 코드를 이해할 수 있다면 이것을 이용해서 만든 스택의 코드는 쉽게 이해할 수 있다. 결국 연결 리스트를 이용한 스택이라고 하더라도 연결 리스트의 링크를 연결하거나 새로운 노드를 생성하는 정도의 코드이기 떄문이다
