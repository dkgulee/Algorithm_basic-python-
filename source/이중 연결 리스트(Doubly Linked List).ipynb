# 이중 연결 리스트

## 이중 연결 리스트의 개념
연결 리스트는 배열과 달리 프로그램의 실행 중에도 동적으로 새로운 노드를 삽입하거나 삭제하기가 간단하며,  물리적인 메모리를 연속적으로 사용하지 않고 링크를 사용하기 떄문에 관리하기가 훨씬 쉽다
단일 연결 리스트는 무조건 한 방향으로만 링크를 따라가야 하기 떄문에 다소 불편한 점이 있다.
**실제로는  링크가 하나더 추가되었다는 점을 제외하고는 그다지 바뀐부분이 없다**

##  이중 연결 리스트의 삽입 알고리즘
이중 연결 리스트는 링크가 하나 더 추가됬기 때문에 삽입과 삭제 알고리즘도 단일 연결 리스트와는 다르다.

![title](../img/1.Doubly_linked_list(insert).png)


1. 연결리스트의 삽입 알고리즘과 같데 next 값만 넣어준다
2. new_node 의 prev설정하고 node_N 의 prev 값을 재설정 해준다

```python
class Node:
    def __init__(self, data, next=None, prev=None):
        self.data = data
        self.next = next
        self.prev = prev
        
    
    def init_list():
        global node_A
        node_A = Node("A")
        node_B = Node("B")
        node_D = Node("D")
        node_E = Node("E")
        node_A.next = node_B
        node_B.prev = node_A
        node_B.next = node_D
        node_D.prev = node_B
        node_D.next = node_E
        node_E.prev = node_D
        
    
    def print_list():
        global node_A
        node = node_A
        while node:
            print(node.data)
            node = node.next
        print
    
    # 노드 삽입하기 
    def insert_node(data):
        global node_A 
        new_node = Node(data)
        node_T = node_A
        node_N = node_A
        
        while node_N.data <= data:
            node_T = node_N
            node_N = node_N.next
        new_node.next = node_N
        node_T.next = new_node
        ## 이중 연결 리스트 삽입 추가 부분
        new_node.prev = node_T
        node_N.prevv = new_node
        
        
    if __name__ == '__main__':
        print("연결 리스트 초기화")
        init_list()
        print_list()
        
        print("<노드 C의 추가>")
        insert_node("C")
        print_list()
```

## 이중 연결 리스트의 삭제 알고리즘

# 이중 연결 리스트 삽입과 삭제

```python
class Node:
    # 생성자 만들어 주기
    def __init__(self,data,next=None,prev=None):
        self.data = data
        self.next = next
        self.prev = prev
        
    # 시작 노드 리스트 만들어 죽기 (node값 next 값 prev 값)
    def init_list():
        global node_A
        node_A = Node("A")
        node_B = Node("B")
        node_D = Node("D")
        node_E = Node("E")
        # 각 노드의 다음 값 넣어주기
        node_A.next = node_B
        node_B.next = node_D
        node_D.next = node_E
        # 각 노드의 전값을 넣어주기
        node_E.prev = node_D
        node_D.prev = node_B
        node_B.prev = node_A
    
    # 이중리스트 노드 삽입하기
    def insert_node(data):
        global node_A
        new_node = Node(data)
        node_T = node_A
        node_N = node_A
        
        while node_N.data <= data:
            node_T = node_N
            node_N = node_N.next
        new_node.next = node_N
        node_T.next = new_node
        node_N.prev = new_node
        new_node.preb = node_T
    
    
    # 이중리스트 노드 삭제하기
    
    def delete_node(del_data):
        global node_A
        node_T = node_A
        node_N = node_T.next
        # 다다음 값 가져오기(단일연결 리스트에서 추가된 소스)
        node_N2 = node_N.next 
        if node_T.data == del_data:
            node_A = node_N
            del node_T
            return
        
        while node_N:
            if node_N.data == del_data:
                node_N2 = node_N.next
                node_T.next = node_N.next
                node_N2.prev = node_N.prev
                del node_N
                break
            node_T = node_N
            node_N = node_N.next
            node_N2 = node_N.next
    
    
    
    
    #리스트를 프린트 해주는 함수 만들기
    def print_list():
        global node_A
        node = node_A
        while node:
            print(node.data)
            node = node.next
    
    if __name__ == "__main__":
        print("<리스트 초기화>")
        init_list()
        print_list()
        print("<C값 삽입하기>")
        insert_node("C")
        print_list()
        print("<D값 삭제하기>")
        delete_node("D")
        print_list()
```
