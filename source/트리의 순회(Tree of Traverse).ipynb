### 1. 트리에서 사용하는 노드를 선언하는 코드


```python
class Node:
    def __init__(self, data):
        self.data =data
        self.left = None
        self.right = None
```

```python
def init_tree():
    global root
    
    new_node = Node("A")
    root = new_node
    new_node = Node("B")
    root.left = new_node
    new_node = Node("C")
    root.right = new_node
    new_node_1 = Node("D")
    new_node_2 = Node("E")
    node = root.left
    node.left = new_node_1
    node.right = new_node_2
    
    new_node_1 = Node("F")
    new_node_2 = Node("G")
    node2 = root.right
    node2.left = new_node_1
    node2.right = new_node_2
    
   
    return  print(root.data, root.left.data, root.right.data, 
                  node.left.data, node.right.data,
                 node2.left.data, node2.right.data ,sep="->")
```

## 전위 순회(Pre-Order Traverse)
트리 구조를 순회하기 위해서 지켜야 할 기본적인 규칙은 **노드는 오직 한번만 방문한다.**
트리 구조에서 트리 구조를 순회하는 방법
1. 가운데 노드를 먼저 방문
2. 왼쪽 노드를 방문
3. 오른쪽 노드를 방문

```python
Image("../img/7-1.PreOrderTraverse.png")
```

```python
def preorder_traverse(node):
    
    # 트리의 끝
    if node == None:
        return
    print(node.data, end = '->')
    # 트리의 끝이 아니면
    preorder_traverse(node.left)
    preorder_traverse(node.right)
    
```

```python
if __name__ == '__main__':
    init_tree()
    print("<preorder_traverse>")
    preorder_traverse(root)
```

## 7.3 중위 순회(In-Order Traverse) 알고리즘

1. 왼쪽 자식 노드를 방문
2. 부모 노드 방문
3. 오른쪽 자식 노드 방문

```python
Image("../img/7-3.In-OrderTraverse.png")
```

```python
def inorder_traverse(node):
    if node == None: return
    inorder_traverse(node.left)
    print(node.data, end='->')
    inorder_traverse(node.right)
```

```python
if __name__ == '__main__':
    init_tree()
    print("<preorder_traverse>")
    preorder_traverse(root)
    print("\n<inorder_traverse>")
    inorder_traverse(root)
```

## 후위 순회(Post-Order Traverse) 알고리즘
1. 왼쪽 자식노드 방문
2. 오른쪽 자식노드 방문
3. 마지막으로 부모 노드 방문

```python
Image("../img/7.4 PostOrderTraverse.png")
```

```python
def postorder_traverse(node):
    if node == None: 
        return
    postorder_traverse(node.left)
    postorder_traverse(node.right)
    print(node.data, end = '->')
```

```python
if __name__ == '__main__':
    init_tree()
    print("<preorder_traverse>")
    preorder_traverse(root)
    print("\n<inorder_traverse>")
    inorder_traverse(root)
    print("\n<postorder_traverse>")
    postorder_traverse(root)
```

## 단계 순회(Level-Order Traverse)알고리즘
단계 순회 알고리즘은 루트 노드부터 단계 순서대로 왼쪽부터 오른쪽으로 방문하는 순회 알고리즘이다.

```python
Image("../img/7-5.Level-OrderTraverse.png")
```

1. 매개 변수로 받은 현재 노드인 node를 큐인 levelq에 저장한다.
2. 큐의 크기가 0이 아니라면, 큐에서 항목을 하나씩 가져와서 그 항목의 데이터를 출력한다.
3. 현재 방문한 노드를 visit_node라 하고 visit_node의 왼쪽 노드가 존재한다면 큐에 visit_node의 왼쪽 노드를 삽입하고, 다시 visit_node의 오른쪽 노드가 존재하는지를 검사한다

```python
levelq = []

def levelorder_traverse(node):
    global levelq
    levelq.append(node)
    while len(levelq) != 0:
        visit_node = levelq.pop(0)
        print(visit_node.data , end = '->')
        
        if visit_node.left != None:
            levelq.append(visit_node.left)
        if visit_node.right != None:
            levelq.append(visit_node.right)
```

```python
if __name__ == '__main__':
    init_tree()
    print("<preorder_traverse>")
    preorder_traverse(root)
    print("\n<inorder_traverse>")
    inorder_traverse(root)
    print("\n<postorder_traverse>")
    postorder_traverse(root)
    print("\n<levelorder_traverse>")
    levelorder_traverse(root)
```

# 정리

#### 이진트리를 순회하는 알고리즘은 전위/중위/후위/단계 순회 알고리즘으로 총 4가지가 있으면, 그 각각의 알고리즘들은 재귀적 호출, 스택이나 큐를 사용하여 구형이 가능하다.

```python
class Node:
    
    def __init__(self,data):
        self.data = data
        self.left = None
        self.right = None
```

```python
def init_tree():
    global root
    root = Node("A")
    root.left = Node("B")
    root.right = Node("C")
        
    node = root.left
    node.left = Node("D")
    node.right = Node("E")
        
    node2 = root.right
    node2.left = Node("F")
    node2.right = Node("G")
        
    return print(root.data, root.left.data, root.right.data,
                 node.data, node.left.data, node.right.data,
                 node2.left.data, node2.right.data, sep='->')

def preorder_traverse(node):
    if node == None:
        return
    print(node.data, end= ' -> ')
    preorder_traverse(node.left)
    preorder_traverse(node.right)
    
def inorder_traverse(node):
    if node == None: return
    inorder_traverse(node.left)
    print(node.data, end='->')
    inorder_traverse(node.right)
    
    
def postorder_traverse(node):
    if node == None: 
        return
    postorder_traverse(node.left)
    postorder_traverse(node.right)
    print(node.data, end = '->')
    
levelq = []

def levelorder_traverse(node):
    global levelq
    levelq.append(node)
    while len(levelq) != 0:
        visit_node = levelq.pop(0)
        print(visit_node.data , end = '->')
        
        if visit_node.left != None:
            levelq.append(visit_node.left)
        if visit_node.right != None:
            levelq.append(visit_node.right)  
```

```python
if __name__ == "__main__":
    init_tree()
    print(root)
    print("<preorder_traverse>")
    preorder_traverse(root)
    print("\n<inorder_traverse>")
    inorder_traverse(root)
    print("\n<postorder_traverse>")
    postorder_traverse(root)
    print("\n<levelorder_traverse>")
    levelorder_traverse(root)
```
